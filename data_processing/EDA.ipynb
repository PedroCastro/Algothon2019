```python
import numpy as np
import pandas as pd
import tqdm
```

```python
supplychain = pd.read_csv("../assets/GSCR_GSPLY.csv")
supplychain[supplychain["customer_exchange"] == "NYSE"]
```

```python
supplychain.sort_values(by="accounting_as_of_date", inplace=True)
```

```python
supplychain.drop(["publish_date", "supplier_name","customer_name", "customer_lei", "reporting_year", "reporting_period", "supplier_lei", "customer_isin", "supplier_isin"], axis=1, inplace=True)
supplychain.dropna(subset=['customer_ticker'], how='all', inplace = True)
supplychain.dropna(subset=['supplier_ticker'], how='all', inplace = True)
supplychain.dropna(subset=['supplier_exchange'], how='all', inplace = True)
supplychain.dropna(subset=['customer_exchange'], how='all', inplace = True)
```

```python
supplychain = supplychain[supplychain["accounting_as_of_date"] > "2000-00-00"]
supplychain["revenue_dependency_annotation"] = supplychain["revenue_dependency_annotation"].astype('category')
```

```python
supplychain
```

```python
supplychain.to_csv("../assets/global_supplychain.csv")
```
